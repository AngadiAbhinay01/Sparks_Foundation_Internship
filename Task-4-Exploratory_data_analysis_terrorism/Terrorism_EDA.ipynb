```python
df = df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region',
                             'provstate':'State','city':'City','attacktype1_txt':'Attack_type','targtype1_txt':'Target_type',
                             'targsubtype1_txt':'Target_subtype','gname':'Gang_name','weaptype1_txt':'Weapon_type'})
```

```python
df = df[['Year','Month','Day','Country','Region','State','City','Attack_type','Target_type','Target_subtype',
         'Gang_name','Weapon_type']]
df.head()
```

# Visualization

```python
weapon_type = df['Weapon_type'].value_counts()
```

```python
plt.figure(figsize=(6,4))
plt.xticks(rotation=90)
plt.bar(weapon_type.index,weapon_type.values)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(20,5));
sns.countplot(x=df.Weapon_type);
```

```python
plt.figure(figsize=(20,5))
year = df['Year'].value_counts()
plt.bar(year.index,year.values)
```

```python
plt.figure(figsize=(20,5));
plt.xticks(rotation = 90)
plt.title('Attacks by Year')
sns.countplot(x=df.Year);
```

```python
plt.figure(figsize=(20,5))
month = df['Month'].value_counts()
plt.bar(month.index,month.values)
```

```python
sns.countplot(x=df['Month'])
```

```python
plt.figure(figsize=(20,5))
plt.title('Attacks by Day')
day = df['Day'].value_counts()
plt.bar(day.index,day.values)
```

```python
plt.figure(figsize=(20,5));
plt.xticks(rotation = 90)
plt.title('Attacks by Country')
country = df['Country'].value_counts()
plt.bar(country.index,country.values);
```

```python
plt.figure(figsize=(28,10));
plt.xticks(rotation = 0)
plt.title('Attacks by Region')
region = df['Region'].value_counts()
plt.bar(region.index,region.values);
```

```python
plt.figure(figsize=(25,5));
plt.rcParams.update({'font.size':12});
plt.xticks(rotation = 0)
plt.title('Based on Attack type')
sns.countplot(x=df.Attack_type);
```
